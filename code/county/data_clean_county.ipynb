```python
employed_industry=pd.read_stata(path+"county_raw/by_industry/employee_by_industry_county.dta")

gva_industry=pd.read_stata(path+"county_raw/by_industry/gva_by_industry_county.dta")
```

```python
employed_weight=employed_industry[["ind_agr","ind_man","ind_ser","Gebietseinheit","year"]]
data_8= pd.merge(
    left=data_7_ready, 
    right=employed_weight,
    on=['Gebietseinheit','year'],
    how='left'
)
data_8=data_8.rename(columns={"ind_agr":"emp_agr","ind_man":"emp_man","ind_ser":"emp_ser"})
data_8
```

```python
gva_weight=gva_industry[["ind_agr","ind_man","ind_ser","Gebietseinheit","year"]]

data_9= pd.merge(
    left=data_8, 
    right=gva_weight,
    on=['Gebietseinheit','year'],
    how='left'
)
data_9=data_9.rename(columns={"ind_agr":"gva_agr","ind_man":"gva_man","ind_ser":"gva_ser"})
data_9
```

```python
data_9=data_9[['year', 'Gebietseinheit', 'EU-Code', 'Regional-key', 'Land','employee',
       'ln_gdp','ln_gva','resident', 'gdp_per_employee', 'volume_of_work',
       'gdp_per_hour_per_employed','firms_no', 'firms_aff_no', 'ratio', 'emp_agr', 'emp_man', 'emp_ser', 'gva_agr', 'gva_man', 'gva_ser']]
```

```python

data_9["emp_agr_weight"]=data_9["emp_agr"]/data_9["employee"]

data_9["emp_man_weight"]=data_9["emp_man"]/data_9["employee"]

data_9["emp_ser_weight"]=data_9["emp_ser"]/data_9["employee"]
data_9["gva_agr_weight"]=data_9["gva_agr"]/(data_9["gva_agr"]+data_9["gva_man"]+data_9["gva_ser"])
data_9["gva_man_weight"]=data_9["gva_man"]/(data_9["gva_agr"]+data_9["gva_man"]+data_9["gva_ser"])
data_9["gva_ser_weight"]=data_9["gva_ser"]/(data_9["gva_agr"]+data_9["gva_man"]+data_9["gva_ser"])
```

```python
data_9['gva_growth_agr'] = np.nan
data_9['gva_growth_man'] = np.nan
data_9['gva_growth_ser'] = np.nan


mask = data_9.groupby('Regional-key')['year'].apply(lambda x: x.shift(1)==x-1) 
#create a mask that tells if the last year existed and just update those years with pct change:
data_9.loc[mask,'gva_growth_agr'] = data_9.groupby('Regional-key')['gva_agr'].pct_change()
data_9.loc[mask,'gva_growth_man'] = data_9.groupby('Regional-key')['gva_man'].pct_change()
data_9.loc[mask,'gva_growth_ser'] = data_9.groupby('Regional-key')['gva_ser'].pct_change()
```

```python
data_9 = data_9.query('2003<year<2013')

data_9
```
