```python
import pandas as pd
import numpy as np
```

```python
path="/Users/effie/Desktop/term3/research_module/data/"
gva_agr=pd.read_excel(path+"county_raw/by_industry/gva_agriculture.xlsx")

gva_agr_back=gva_agr[['EU-Code','Regional-key','Land',"Gebietseinheit"]]
gva_agr_back
```

```python
gva_agr=gva_agr.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])
gva_agr.dropna()
gva_agr=gva_agr.set_index('Gebietseinheit')

gva_agr_unstack=gva_agr.unstack()
gva_agr_unstack=pd.DataFrame(gva_agr_unstack)
gva_agr_unstack=gva_agr_unstack.reset_index(level=[0,1])

gva_agr_unstack= gva_agr_unstack.rename(columns={'level_0': 'year', 0: 'ind_agr'})

gva_agr_unstack=gva_agr_unstack.dropna()
gva_agr_unstack
```

```python
data_1= pd.merge(
    left=gva_agr_unstack, 
    right=gva_agr_back,
    on=['Gebietseinheit'],
    how='left'
)
data_1
```

```python
gva_man=pd.read_excel(path+"county_raw/by_industry/gva_manufacturing.xlsx")
gva_man
gva_man=gva_man.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])

gva_man=gva_man.set_index('Gebietseinheit')

gva_man_unstack=gva_man.unstack()
gva_man_unstack=pd.DataFrame(gva_man_unstack)
gva_man_unstack=gva_man_unstack.reset_index(level=[0,1])

gva_man_unstack= gva_man_unstack.rename(columns={'level_0': 'year', 0: 'ind_man'})
gva_man_unstack=gva_man_unstack.dropna()
gva_man_unstack
```

```python
data_2= pd.merge(
    left=data_1, 
    right=gva_man_unstack,
    on=['Gebietseinheit',"year"],
    how='left'
)
data_2
```

```python
gva_ser=pd.read_excel(path+"county_raw/by_industry/gva_service.xlsx")
gva_ser=gva_ser.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])

gva_ser=gva_ser.set_index('Gebietseinheit')

gva_ser_unstack=gva_ser.unstack()
gva_ser_unstack=pd.DataFrame(gva_ser_unstack)
gva_ser_unstack=gva_ser_unstack.reset_index(level=[0,1])

gva_ser_unstack= gva_ser_unstack.rename(columns={'level_0': 'year', 0: 'ind_ser'})
gva_ser_unstack
```

```python
data_3= pd.merge(
    left=data_2, 
    right=gva_ser_unstack,
    on=['Gebietseinheit',"year"],
    how='left'
)
data_3
```

```python
gva_fin=pd.read_excel(path+"county_raw/by_industry/gva_finance.xlsx")
gva_fin=gva_fin.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])

gva_fin=gva_fin.set_index('Gebietseinheit')

gva_fin_unstack=gva_fin.unstack()
gva_fin_unstack=pd.DataFrame(gva_fin_unstack)
gva_fin_unstack=gva_fin_unstack.reset_index(level=[0,1])

gva_fin_unstack= gva_fin_unstack.rename(columns={'level_0': 'year', 0: 'ind_ser_fin'})
gva_fin_unstack
```

```python
data_4= pd.merge(
    left=data_3, 
    right=gva_fin_unstack,
    on=['Gebietseinheit',"year"],
    how='left'
)
data_4
```

```python
gva_trade=pd.read_excel(path+"county_raw/by_industry/gva_trade.xlsx")
gva_trade=gva_trade.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])

gva_trade=gva_trade.set_index('Gebietseinheit')

gva_trade_unstack=gva_trade.unstack()
gva_trade_unstack=pd.DataFrame(gva_trade_unstack)
gva_trade_unstack=gva_trade_unstack.reset_index(level=[0,1])

gva_trade_unstack= gva_trade_unstack.rename(columns={'level_0': 'year', 0: 'ind_ser_trade'})
gva_trade_unstack
```

```python
data_5= pd.merge(
    left=data_4, 
    right=gva_trade_unstack,
    on=['Gebietseinheit',"year"],
    how='left'
)
data_5
```

```python
gva_public=pd.read_excel(path+"county_raw/by_industry/gva_public.xlsx")
gva_public=gva_public.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])

gva_public=gva_public.set_index('Gebietseinheit')

gva_public_unstack=gva_public.unstack()
gva_public_unstack=pd.DataFrame(gva_public_unstack)
gva_public_unstack=gva_public_unstack.reset_index(level=[0,1])

gva_public_unstack= gva_public_unstack.rename(columns={'level_0': 'year', 0: 'ind_ser_public'})
gva_public_unstack
```

```python
data_6= pd.merge(
    left=data_5, 
    right=gva_public_unstack,
    on=['Gebietseinheit',"year"],
    how='left'
)
data_6
```

```python

data_6.loc[data_6['Land'] =="HH", 'Regional-key'] = "02000"
data_6.loc[data_6['Land'] =="BE", 'Regional-key'] = "11000"
```

```python
data_6 = data_6.replace('.', np.nan)
data_6= data_6.replace('...', np.nan)
data_6['Regional-key']=data_6['Regional-key'].astype(str)
data_6["ind_agr"] = pd.to_numeric(data_6["ind_agr"], downcast="float")

data_6=data_6[data_6['Regional-key'].str.len()>=3]
data_6['Regional-key']= np.where(data_6['Regional-key'].str.len() < 5, "0"+data_6['Regional-key'],data_6['Regional-key'])

data_6=data_6[['year', 'Gebietseinheit',  'EU-Code', 'Regional-key', 'Land','ind_agr',
       'ind_man', 'ind_ser',"ind_ser_fin","ind_ser_trade","ind_ser_public"]]
data_6
```

```python
data_6.to_stata(path+'county_raw/by_industry/gva_by_industry_county.dta')
```
