```python
import pandas as pd
import numpy as np
```

```python
path="/Users/effie/Desktop/term3/research_module/data/"
data=pd.read_csv(path+"firm_clean/key for county identification.csv")
# data=pd.read_stata("/Users/mengxi/Desktop/financial research/HAN/total_data_county.dta")
data=pd.DataFrame(data)
data=data.reset_index().drop(['index'], axis=1)
data.index=data.index+1
```

```python
data['ort_1'] = data['Gebietseinheit'].str.replace(", Kreisfreie Stadt", "", n=-1)
data['ort_2'] = data['ort_1'].str.replace(", Regionalverband", "", n=-1)
data['ort_3'] = data['ort_2'].str.replace(", Landeshauptstadt, Stadtkreis", "", n=-1)
data['ort_4'] = data['ort_3'].str.replace(", Universitätsstadt, Stadtkreis", "", n=-1)
data['ort_5'] = data['ort_4'].str.replace(", Landeshauptstadt Städte", "", n=-1)
data['ort_6'] = data['ort_5'].str.replace(", Wissenschaftsstadt Städte", "", n=-1)
data['ort_7'] = data['ort_6'].str.replace(", Hansestadt Städte", "", n=-1)
data['ort_8'] = data['ort_7'].str.replace(", Stadtkreis", "", n=-1)
data['ort_9'] = data['ort_8'].str.replace(", Kreis", "", n=-1)
data['ort_10'] = data['ort_9'].str.replace(", documenta-Stadt Städte", "", n=-1)
data['ort_11'] = data['ort_10'].str.replace(", NUTS 2-Region", "", n=-1)
data['ort_'] = data['ort_11'].str.replace(", Landkreis", "", n=-1)
```

```python
data=data[['year', 'Gebietseinheit', 'employee', 'EU_Code',
       'Regional_key', 'Land', 'ln_gdp', 'resident', 'ln_gva',
       'gdp_per_employee', 'volume_of_work', 'gdp_per_hour_per_employed',
       'ort_']]
```
