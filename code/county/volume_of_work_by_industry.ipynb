```python
import pandas as pd
import numpy as np
```

```python
path="/Users/effie/Desktop/term3/research_module/data/"
vow_agr=pd.read_excel(path+"county_raw/by_industry/vow_agriculture.xlsx")

vow_agr_back=vow_agr[['EU-Code','Regional-key','Land',"Gebietseinheit"]]
vow_agr_back
```

```python
vow_agr=vow_agr.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])
vow_agr.dropna()
vow_agr=vow_agr.set_index('Gebietseinheit')

vow_agr_unstack=vow_agr.unstack()
vow_agr_unstack=pd.DataFrame(vow_agr_unstack)
vow_agr_unstack=vow_agr_unstack.reset_index(level=[0,1])

vow_agr_unstack= vow_agr_unstack.rename(columns={'level_0': 'year', 0: 'ind_agr'})

vow_agr_unstack=vow_agr_unstack.dropna()
vow_agr_unstack
```

```python
data_1= pd.merge(
    left=vow_agr_unstack, 
    right=vow_agr_back,
    on=['Gebietseinheit'],
    how='left'
)
data_1
```

```python
vow_man=pd.read_excel(path+"county_raw/by_industry/vow_manufacturing.xlsx")
vow_man
vow_man=vow_man.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])

vow_man=vow_man.set_index('Gebietseinheit')

vow_man_unstack=vow_man.unstack()
vow_man_unstack=pd.DataFrame(vow_man_unstack)
vow_man_unstack=vow_man_unstack.reset_index(level=[0,1])

vow_man_unstack= vow_man_unstack.rename(columns={'level_0': 'year', 0: 'ind_man'})
vow_man_unstack=vow_man_unstack.dropna()
vow_man_unstack
```

```python
data_2= pd.merge(
    left=data_1, 
    right=vow_man_unstack,
    on=['Gebietseinheit',"year"],
    how='left'
)
data_2
```

```python
vow_ser=pd.read_excel(path+"county_raw/by_industry/vow_service.xlsx")
vow_ser=vow_ser.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])

vow_ser=vow_ser.set_index('Gebietseinheit')

vow_ser_unstack=vow_ser.unstack()
vow_ser_unstack=pd.DataFrame(vow_ser_unstack)
vow_ser_unstack=vow_ser_unstack.reset_index(level=[0,1])

vow_ser_unstack= vow_ser_unstack.rename(columns={'level_0': 'year', 0: 'ind_ser'})
vow_ser_unstack
```

```python
data_3= pd.merge(
    left=data_2, 
    right=vow_ser_unstack,
    on=['Gebietseinheit',"year"],
    how='left'
)
data_3
```

```python
vow_fin=pd.read_excel(path+"county_raw/by_industry/vow_finance.xlsx")
vow_fin=vow_fin.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])

vow_fin=vow_fin.set_index('Gebietseinheit')

vow_fin_unstack=vow_fin.unstack()
vow_fin_unstack=pd.DataFrame(vow_fin_unstack)
vow_fin_unstack=vow_fin_unstack.reset_index(level=[0,1])

vow_fin_unstack= vow_fin_unstack.rename(columns={'level_0': 'year', 0: 'ind_ser_fin'})
vow_fin_unstack
```

```python
data_4= pd.merge(
    left=data_3, 
    right=vow_fin_unstack,
    on=['Gebietseinheit',"year"],
    how='left'
)
data_4
```

```python
vow_trade=pd.read_excel(path+"county_raw/by_industry/vow_trade.xlsx")
vow_trade=vow_trade.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])

vow_trade=vow_trade.set_index('Gebietseinheit')

vow_trade_unstack=vow_trade.unstack()
vow_trade_unstack=pd.DataFrame(vow_trade_unstack)
vow_trade_unstack=vow_trade_unstack.reset_index(level=[0,1])

vow_trade_unstack= vow_trade_unstack.rename(columns={'level_0': 'year', 0: 'ind_ser_trade'})
vow_trade_unstack
```

```python
data_5= pd.merge(
    left=data_4, 
    right=vow_trade_unstack,
    on=['Gebietseinheit',"year"],
    how='left'
)
data_5
```

```python
vow_public=pd.read_excel(path+"county_raw/by_industry/vow_public.xlsx")
vow_public=vow_public.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])

vow_public=vow_public.set_index('Gebietseinheit')

vow_public_unstack=vow_public.unstack()
vow_public_unstack=pd.DataFrame(vow_public_unstack)
vow_public_unstack=vow_public_unstack.reset_index(level=[0,1])

vow_public_unstack= vow_public_unstack.rename(columns={'level_0': 'year', 0: 'ind_ser_public'})
vow_public_unstack
```

```python
data_6= pd.merge(
    left=data_5, 
    right=vow_public_unstack,
    on=['Gebietseinheit',"year"],
    how='left'
)
data_6
```

```python
data_6.loc[data_6['Land'] =="HH", 'Regional-key'] = "02000"
data_6.loc[data_6['Land'] =="BE", 'Regional-key'] = "11000"
```

```python
data_6 = data_6.replace('.', np.nan)
data_6= data_6.replace('...', np.nan)
data_6['Regional-key']=data_6['Regional-key'].astype(str)
data_6["ind_agr"] = pd.to_numeric(data_6["ind_agr"], downcast="float")

data_6=data_6[data_6['Regional-key'].str.len()>=3]
data_6['Regional-key']= np.where(data_6['Regional-key'].str.len() < 5, "0"+data_6['Regional-key'],data_6['Regional-key'])

data_6=data_6[['year', 'Gebietseinheit',  'EU-Code', 'Regional-key', 'Land','ind_agr',
       'ind_man', 'ind_ser',"ind_ser_fin","ind_ser_trade","ind_ser_public"]]
data_6
```

```python
data_6.to_stata(path+'county_raw/by_industry/vow_by_industry_county.dta')
```
