```python
import pandas as pd
import numpy as np
```

```python
path="/Users/effie/Desktop/term3/research_module/data/"
employed_agr=pd.read_excel(path+"county_raw/by_industry/employed_agriculture.xlsx")

employed_agr_back=employed_agr[['EU-Code','Regional-key','Land',"Gebietseinheit"]]
employed_agr_back
```

```python
employed_agr
```

```python
employed_agr=employed_agr.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])

employed_agr=employed_agr.set_index('Gebietseinheit')

employed_agr_unstack=employed_agr.unstack()
employed_agr_unstack=pd.DataFrame(employed_agr_unstack)
employed_agr_unstack=employed_agr_unstack.reset_index(level=[0,1])

employed_agr_unstack= employed_agr_unstack.rename(columns={'level_0': 'year', 0: 'ind_agr'})
employed_agr_unstack
```

```python
data_1= pd.merge(
    left=employed_agr_unstack, 
    right=employed_agr_back,
    on=['Gebietseinheit'],
    how='left'
)
data_1
```

```python
employed_man=pd.read_excel(path+"county_raw/by_industry/employed_manufacturing.xlsx")
# employed_agr_back=employed_agr[['EU-Code','Regional-key','Land',"Gebietseinheit"]]
employed_man
employed_man=employed_man.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])

employed_man=employed_man.set_index('Gebietseinheit')

employed_man_unstack=employed_man.unstack()
employed_man_unstack=pd.DataFrame(employed_man_unstack)
employed_man_unstack=employed_man_unstack.reset_index(level=[0,1])

employed_man_unstack= employed_man_unstack.rename(columns={'level_0': 'year', 0: 'ind_man'})
employed_man_unstack
```

```python
data_2= pd.merge(
    left=data_1, 
    right=employed_man_unstack,
    on=['Gebietseinheit',"year"],
    how='left'
)
data_2
```

```python
employed_ser=pd.read_excel(path+"county_raw/by_industry/employed_service.xlsx")
employed_ser=employed_ser.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land',"NUTS 1","NUTS 2","NUTS 3"])

employed_ser=employed_ser.set_index('Gebietseinheit')

employed_ser_unstack=employed_ser.unstack()
employed_ser_unstack=pd.DataFrame(employed_ser_unstack)
employed_ser_unstack=employed_ser_unstack.reset_index(level=[0,1])

employed_ser_unstack= employed_ser_unstack.rename(columns={'level_0': 'year', 0: 'ind_ser'})
employed_ser_unstack
```

```python
data_3= pd.merge(
    left=data_2, 
    right=employed_ser_unstack,
    on=['Gebietseinheit',"year"],
    how='left'
)
data_3
```

```python
employed_fin=pd.read_excel(path+"county_raw/by_industry/employed_finance.xlsx")
employed_fin
employed_fin=employed_fin.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land'])

employed_fin=employed_fin.set_index('Gebietseinheit')

employed_fin_unstack=employed_fin.unstack()
employed_fin_unstack=pd.DataFrame(employed_fin_unstack)
employed_fin_unstack=employed_fin_unstack.reset_index(level=[0,1])

employed_fin_unstack= employed_fin_unstack.rename(columns={'level_0': 'year', 0: 'ind_ser_fin'})
employed_fin_unstack
```

```python
data_4= pd.merge(
    left=data_3, 
    right=employed_fin_unstack,
    on=['Gebietseinheit',"year"],
    how='left'
)
data_4
```

```python
employed_trade=pd.read_excel(path+"county_raw/by_industry/employed_trade.xlsx")
employed_trade=employed_trade.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land'])
employed_trade=employed_trade.set_index('Gebietseinheit')
employed_trade_unstack=employed_trade.unstack()
employed_trade_unstack=pd.DataFrame(employed_trade_unstack)
employed_trade_unstack=employed_trade_unstack.reset_index(level=[0,1])
employed_trade_unstack= employed_trade_unstack.rename(columns={'level_0': 'year', 0: 'ind_ser_trade'})
employed_trade_unstack
```

```python
employed_public=pd.read_excel(path+"county_raw/by_industry/employed_public.xlsx")
employed_public=employed_public.drop(columns=["Lfd. Nr.",'EU-Code','Regional-key','Land'])
employed_public=employed_public.set_index('Gebietseinheit')
employed_public_unstack=employed_public.unstack()
employed_public_unstack=pd.DataFrame(employed_public_unstack)
employed_public_unstack=employed_public_unstack.reset_index(level=[0,1])
employed_public_unstack= employed_public_unstack.rename(columns={'level_0': 'year', 0: 'ind_ser_public'})
employed_public_unstack
```

```python
data_6= pd.merge(
    left=data_5, 
    right=employed_public_unstack,
    on=['Gebietseinheit',"year"],
    how='left'
)
data_6
```

```python

data_6.loc[data_6['Land'] =="HH", 'Regional-key'] = "02000"
data_6.loc[data_6['Land'] =="BE", 'Regional-key'] = "11000"

```

```python
data_6 = data_6.replace('.', np.nan)
data_6= data_6.replace('...', np.nan)
data_6['Regional-key']=data_6['Regional-key'].astype(str)
data_6=data_6[data_6['Regional-key'].str.len()>=3]
data_6['Regional-key']= np.where(data_6['Regional-key'].str.len() < 5, "0"+data_6['Regional-key'],data_6['Regional-key'])

data_6=data_6[['year', 'Gebietseinheit',  'EU-Code', 'Regional-key', 'Land','ind_agr',
       'ind_man', 'ind_ser',"ind_ser_fin","ind_ser_trade","ind_ser_public"]]
data_6
```
