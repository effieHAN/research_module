```python
path="/Users/effie/Desktop/term3/research_module/data/"
data=pd.read_stata(path+"county_clean/total_and_by_industry_data_county.dta")

data=data.drop(columns=['index'])

data["res_ln"]=np.log(data["resident"]*1000)
data["emp_ln"]=np.log(data["employee"]*1000)
data["emp_agr_ln"]=np.log(data["emp_agr"]*1000)
data["emp_man_ln"]=np.log(data["emp_man"]*1000)
data["emp_ser_ln"]=np.log(data["emp_ser"]*1000)

data
```

### OLS

```python
exog_vars = ['affected_after','county_affected','year_post',"res_ln",'gva_agr_weight','gva_man_weight']
exog = sm.add_constant(df_fe[exog_vars])
mod = PanelOLS(df_fe.ln_gdp,exog,entity_effects=False)
res_fe_1 = mod.fit(cov_type = 'clustered')

exog_vars = ['affected_after','county_affected','year_post',"res_ln",'gva_agr_weight','gva_man_weight']
exog = sm.add_constant(df_fe[exog_vars])
mod = PanelOLS(df_fe.ln_gva,exog,entity_effects=False)
res_fe_2 = mod.fit(cov_type = 'clustered')


exog_vars = ['affected_after','county_affected','year_post',"res_ln",'gva_agr_weight']
exog = sm.add_constant(df_fe[exog_vars])
mod = PanelOLS(df_fe.gva_growth_agr,exog,entity_effects=False)
res_fe_3 = mod.fit(cov_type = 'clustered')


exog_vars = ['affected_after','county_affected','year_post',"res_ln",'gva_man_weight']
exog = sm.add_constant(df_fe[exog_vars])
mod = PanelOLS(df_fe.gva_growth_man,exog,entity_effects=False)
res_fe_4 = mod.fit(cov_type = 'clustered')

exog_vars = ['affected_after','county_affected','year_post',"res_ln", 'gva_ser_weight']
exog = sm.add_constant(df_fe[exog_vars])
mod = PanelOLS(df_fe.gva_growth_man,exog,entity_effects=False)
res_fe_5 = mod.fit(cov_type = 'clustered')



print(compare({"(1)Log_gdp": res_fe_1, "(3)Log_gva": res_fe_2,"(3)Agr_gva": res_fe_3, "(4)Man_gva": res_fe_4, "(5)Ser_gva": res_fe_5}, stars = True, precision = "std_errors"))
```

### FIXED Effect

```python
#for total gdp
mod_total_gdp = lm.PanelOLS.from_formula('ln_gdp ~ affected_after + res_ln + EntityEffects +TimeEffects +gva_agr_weight + gva_man_weight',df_fe)
total_gdp = mod_total_gdp.fit(cov_type = 'clustered')
#for total gva
mod_total_gva = lm.PanelOLS.from_formula('ln_gva ~ affected_after + res_ln + EntityEffects +TimeEffects +gva_agr_weight + gva_man_weight ',df_fe)
total_gva = mod_total_gdp.fit(cov_type = 'clustered')

#for agr gva
mod_gva_agr = lm.PanelOLS.from_formula('gva_growth_agr ~ affected_after + res_ln + EntityEffects +TimeEffects +gva_agr_weight ',df_fe)
agr_gva = mod_gva_agr.fit(cov_type = 'clustered')
#for man gva
mod_gva_man = lm.PanelOLS.from_formula('gva_growth_man ~ affected_after + res_ln + EntityEffects +TimeEffects  + gva_man_weight ',df_fe)
man_gva = mod_gva_man.fit(cov_type = 'clustered')
#for ser gva
mod_gva_ser = lm.PanelOLS.from_formula('gva_growth_ser ~ affected_after +res_ln+  EntityEffects +TimeEffects + gva_ser_weight ',df_fe)
ser_gva = mod_gva_ser.fit(cov_type = 'clustered')
print(compare({"(1)Log_gdp": total_gdp,"(2)Log_gva": total_gva, "(3)Log_gva_agr": agr_gva,"(4)Log_gva_man": man_gva,"(5)Log_gva_ser": ser_gva}, stars = True, precision = "std_errors"))
```
