```python
employed_industry=pd.read_stata("/Users/effie/Desktop/term3/research_module/data/clean/employee_by_industry_county.dta")

gva_industry=pd.read_stata("/Users/effie/Desktop/term3/research_module/data/clean/employee_by_industry_county.dta")
```

```python
employed_industry
```

```python
employed_industry["emp_agr_weight"]=employed_industry["ind_agr"]/np.sum(employed_industry["ind_agr"]+employed_industry["ind_man"]+employed_industry["ind_ser"])
employed_industry["emp_man_weight"]=employed_industry["ind_man"]/np.sum(employed_industry["ind_agr"]+employed_industry["ind_man"]+employed_industry["ind_ser"])
employed_industry["emp_ser_weight"]=employed_industry["ind_ser"]/np.sum(employed_industry["ind_agr"]+employed_industry["ind_man"]+employed_industry["ind_ser"])
```

```python
employed_industry
```

```python
employed_weight=employed_industry[["emp_agr_weight","emp_man_weight","emp_ser_weight","Gebietseinheit","year"]]
employed_weight
```

```python
data_8= pd.merge(
    left=data_7, 
    right=employed_weight,
    on=['Gebietseinheit','year'],
    how='left'
)
data_8
```

```python
gva_industry["gva_agr_weight"]=gva_industry["ind_agr"]/np.sum(gva_industry["ind_agr"]+gva_industry["ind_man"]+gva_industry["ind_ser"])
gva_industry["gva_man_weight"]=gva_industry["ind_man"]/np.sum(gva_industry["ind_agr"]+gva_industry["ind_man"]+gva_industry["ind_ser"])
gva_industry["gva_ser_weight"]=gva_industry["ind_ser"]/np.sum(gva_industry["ind_agr"]+gva_industry["ind_man"]+gva_industry["ind_ser"])
gva_weight=gva_industry[["gva_agr_weight","gva_man_weight","gva_ser_weight","Gebietseinheit","year"]]

data_9= pd.merge(
    left=data_8, 
    right=gva_weight,
    on=['Gebietseinheit','year'],
    how='left'
)
data_9
```

```python
data_9.columns
```

```python
data_9=data_9[['year', 'Gebietseinheit', 'EU-Code', 'Regional-key', 'Land','employee',
       'ln_gdp','ln_gva', 'gdp_per_employee', 'volume_of_work',
       'gdp_per_hour_per_employed', 'resident',  'emp_agr_weight', 'emp_man_weight',
       'emp_ser_weight', 'gva_agr_weight', 'gva_man_weight', 'gva_ser_weight']]
data_9
```

```python
data_9.to_stata('../data/clean/total_data_county_with_industry_weight.dta')
```
