```python
import pandas as pd 
import numpy as np
```

```python
data=pd.read_stata("/Users/mengxi/Desktop/firmbank_topic3.dta")
```

```python
data
```

```python
query_data = data.query('empl<20')
data.loc[:,'small'] = data.index.map(lambda x : 1 if x in query_data.index else 0)

query_data = data.query('year>2008')
data.loc[:,'year_post'] = data.index.map(lambda x : 1 if x in query_data.index else 0)


data
```

```python
data.columns
#'shfd' is equity 
```

```python
merged_data=data[['BvDID_firm', 'year','NamederBank','BvDID_bank','bank_type','postcode_bank','empl','toas','cash','shfd','gros','year_post','small']]
merged_data
# 数据本身有错或者还需要完善，比如下表最后一行，bank_type应该是Savings banks，但是数据给出的是NaN
```

```python
merged_data = merged_data.query('bank_type=="Savings banks"')
merged_data = merged_data.query('2004<year<2013')
merged_data 
```

```python
merged_data=merged_data.rename(columns={"BvDID_firm": "firm_id", "NamederBank": "bank_name", "BvDID_bank": "bank_id",
                            "empl": "no_of_employee","cash":'total_cash',"shfd": "total_capital","gros":'gros_profit',"toas":'total_assets'}, errors="raise")
```

```python
merged_data=merged_data.reset_index().drop(['index'], axis=1)
merged_data.index=merged_data.index+1
merged_data
```

```python
merged_data.loc[:,'ln_no_emp'] = np.log(merged_data.no_of_employee)
merged_data.loc[:,'ln_assets'] = np.log(merged_data.total_assets)
```

```python
merged_data.to_csv(path_or_buf=r'/Users/mengxi/Desktop/financial research/04-13_data_raw.csv')
```
